\section{Motivation}

 I wrestled throughout my childhood and high school. I wasn't great, definitely not the strongest or fastest: my father made me do it. Because of this, I would try and do research on my opponents style, favorite moves, strengths, and weaknesses while they were wrestling in their match before me. Nearly all competitions were tournament style, so the further you went the more important the next match was. I found it interesting different teams had different approaches to 'scoping-out-the-competition.' Some coaches believed that it was a distraction to a warm up or something that would get in their athlete's heads and thus wouldn't allow their athletes to do so. Other coaches, like many of mine, would organize all the match locations of our future competitors and would let us know when one of our competitors was about to compete so we could scope out our competition. I always thought it was silly of other other coaches not to allow their players to learn from their competitor's matches. However, I have seen examples where seeing your opponents match turns out bad. I had teammates that easily got in their head, even if they were really good wrestlers. They would see an opponent preform well and immediately become nervous and tell me how they expected to lose. There was clearly a certain type of people that could easily 'psyche-themselves-out'. From another prospective, I had a set of moves that were clearly my strongest. Even if I was expecting to win the match I would prefer to strengthen my lead or end the match faster by using these moves. If I saw my next opponent on the side of the mat as I approached the referee, I would want to save these moves so that my opponent could not practice defences, change their style, or expect me to use these moves. It was clear who your opponent was as everyone wore team gear and there was only one competitor per weight class per team. I also could tell that sometimes when I had shown my strengths my next opponent had  become very aware and changed their style and defense to disadvantage my strengths. However, sometimes it was also clear when my opponents had 'psyched-themselves-out': there was no strength, or commitment in any of the moves they tried, making it easier to win the match. Once again, as these were tournament style competitions and people were seeded, the next match was typically more important and difficult to win than the current.
    
        $\tab$
    
 I am interested in this game from both the prospective of a watching competitor and a competing competitor. Should the viewing competitor spectate the match, knowing whether or not they are the type to 'psyche-themselves-out'? Should a wrestler about to wrestle use there best moves if they see their future competitor on the side of the mat? This logic can even apply other games where future plays might be important that the current and you may or may not want to show your best play yet, especially one-on-one games where it is easy to get intimidated. This game is further explored in this game. 



\section{Build}
 A model as shown below represents our game. It is said that nature makes the decision of whether a viewer is easily intimidated or confident and the viewer knows their type. From experience, it makes sense that nature make the decision that 30\% of competitors/viewers are easily intimidated and the other 70\% are confident; this is known by both players. The viewer then has the choice to watch their next opponent's match. From there, the current wrestler can then decide to show their strength and best moves or not. The current wrestler knows if their next opponent is watching or not but not whether they are easily intimidated of confident. 
    
        $\tab$
    
The payoffs are decided from the current and next match. If the viewer, or next competitor, does not watch the match, they earn no payoff. The wrestler using their strengths in the first match is always worth 1 point to them as they get more team points, more individual points, and might even get to showoff to their friends and family watching. Effects that carry into the next match are both potentially worth more points as the next match is more worth more and a zero-sum-game as any one advantage to one wrestler is a disadvantage to the other. If the the viewer watches the match but the wrestler does not show their strengths they gain 1 point that the wrestler loses. This makes sense as they see some basic techniques of the wrestler but nothing that could intimidate an easily intimidated viewer. If a confident viewer watches the match and the wrestler shows their strengths, the viewer gains 3 points that the wrestler loses as they learned lots from the match and have a large advantage in the next match; however, the wrestler still gained a point from their current match. If an easily intimidated viewer watches the match and the wrestler shows their strengths the wrestler gains 2 points that the viewer loses from successfully intimidating the viewer as this is more powerful than what the viewer may have learned. The wrestler again gains 1 point from using their best moves in the current match. This is all easily viewed in the figure below.

\section{Solve}
Lets consider if both types of players pooling on watching the match is a perfect Bayesian Nash equilibrium. In this case, the viewers move of deciding to watch is uninformative to the wrestler. 
\begin{align*}
P(E|W) = P(E) = 0.3\\
P(C|W) = P(C) = 0.7
\end{align*}

Writing down the expected payoff from each move can help determine the wrestlers best response. 

\begin{align*}
    E(\pi_{Wr}(S|W))&=P(E|W)*\pi_{Wr}(E,W,S) + P(C|W)*\pi_{Wr}(C,W,S) \\
    E(\pi_{Wr}(S|W))&=0.3 (3) + 0.7(-2)\\
    E(\pi_{Wr}(S|W))&=-0.5
\end{align*}
\begin{align*}
    E(\pi_{Wr}(N|W))&=P(E|W)*\pi_{Wr}(E,W,N) + P(C|W)*\pi_{Wr}(C,W,N) \\
    E(\pi_{Wr}(N|W))&=0.3 (-1) + 0.7(-1)\\
    E(\pi_{Wr}(N|W))&=-1
\end{align*}
The wrestlers best response is to show their best moves as the expected payoff of that choice is higher than the expected payoff of not showing their best moves. It is no longer the easily intimidated viewer's bests response to watch. This holds true no matter what assumptions the wrestler might make if they realized their competitor was not watching as the viewers payoff will be 0, no matter. This is therefore not a perfect Bayesian Nash equilibrium.

It is clear that both types of players pooling on not watching the match is not a perfect Bayesian Nash equilibrium because watching the match is a dominant strategy for a confident viewer.

In that case, lets consider a separating outcome that is a perfect Bayesian Nash equilibrium where a confident viewer always watches the match and an easily intimidated viewer never watched the match. In that case, it is the wrestlers best response to not show their best moves if they see their next opponent watching the match and to show their best moves if they see that their next opponent is not watching the match. Given these best responses, it is the best response of the easily intimidated viewer to change their decision and watch the match; therefore, this is not a perfect Bayesian Nash equilibrium.

Since there is no pure strategy perfect Bayesian Nash, lets consider a mixed strategy outcome. It is a confident viewer's dominant strategy to always watch the match so lets consider a strategy where they do this and an easily intimidated viewer sometimes watches the match: with a probability $p$. In addition, the wrestler shows their strengths when their opponent is watching with a probability $q$. We will solve to see if there exists a $p$ and $q$ that satisfy a perfect Bayesian Equilibrium. It is clear that is is also the wrestlers dominant strategy to show their strengths is their future opponent is not watching. This will be reviewed more in the Enhance Section. Therefore;
\begin{align*}
    P(W|E) &= p\\
    P(S|W) &= q\\
    P(E|NW) &= 1\\
    P(W|C) &= 1\\
\end{align*}

Solving for the probability that the viewer is easily intimidated given that they are watching will allow us to create a best response for the wrestler when their opponent is watching. This can be done with Bayes Theorem:
\begin{align*}
    P(E|W)&=\frac{P(W|E)+P(E)}{P(W)}\\
    P(E|W)&=\frac{P(W|E)+P(E)}{P(W|E)P(E)+P(W|C)P(C)}\\
    P(E|W)&=\frac{.3p}{.3p+.7}\\
    P(E|W)&=\frac{3p}{3p+7}
\end{align*}

The probably that the viewer is instead confident is therefore:
\begin{align*}
    P(C|W)&= 1-P(E|W)\\
    P(C|W)&=1-\frac{3p}{3p+7}\\
    P(C|W)&=\frac{7}{3p+7}
\end{align*}
The wrestler wants to show their strengths with a probability that maximizes their expected payoff. 
\begin{align*}
   E(\pi_{Wr,W})= &P(S|W)P(E|W)\pi_{Wr}(E,W,S)\\
                 + &P(S|W)P(C|W)\pi_{Wr}(C,W,S)\\
                 + &P(N|W)P(E|W)\pi_{Wr}(E,W,N)\\
                 + &P(N|W)P(C|W)\pi_{Wr}(C,W,N)\\
    E(\pi_{Wr,W})= &q(\frac{3p}{3p+7})(3) \\
                + &q(\frac{7}{3p+7})(-2) \\
                + &(1-q)(\frac{3p}{3p+7})(-1) \\
                + &(1-q)(\frac{7}{3p+7})(-1)\\
    &\max_q{q(\frac{9p-14}{3p+7}) + (1-q)(-1)}\\
    &\max_q{q\underbrace{(\frac{9p-14}{3p+7}+1)} -1}\\
\end{align*}
Solving for when the under-braced part is positive, negative, or zero will tell us what '$q$' the wrestler would choose to maximize their payoff. 
\begin{align*}
    \frac{9p-14}{3p+7}+1&=0\\
    \frac{9p-14}{3p+7}&=-1\\
    9p-14 &= -3p-7\\
    12p&=7\\
    p &= \frac{7}{12}
\end{align*}
Therefore, the wrestler chooses:
\begin{align*}
q* = 0\ if\ p<\frac{7}{12}\\
q* \isin (0,1)\ if\ p=\frac{7}{12}\\
q* = 1\ if\ p>\frac{7}{12}
\end{align*}
Our previous analysis of pure strategies showed that one is not part of a perfect Bayesian equilibrium. Therefore, $p*=\frac{7}{12}$ must be true if this mixed strategy is part of a perfect Bayesian equilibrium. 

A best response should also be determined for the viewer that is easily intimidated. The viewer wants to watch the match with a probability that maximizes their expected payoff. 
\begin{align*}
   E(\pi_{V,E})= &P(S|W)P(W|E)\pi_{V}(E,W,S)\\
                 + &P(S|NW)P(NW|E)\pi_{V}(E,NW,S)\\
                 + &P(N|W)P(W|E)\pi_{V}(E,W,N)\\
                 + &P(N|NW)P(NW|E)\pi_{V}(E,NW,N)\\
    E(\pi_{V,E})= &qp(-2) \\
                + &1(1-p)(0) \\
                + &(1-q)p(1) \\
                + &(0)(1-p)(0)\\
    &\max_q{p(-2q) + p -pq}\\
    &\max_q{p\underbrace{(-3q+1)}}\\
\end{align*}
Solving for when the under-braced part is positive, negative, or zero will tell us what '$p$' the viewer would choose to maximize their payoff. 
\begin{align*}
    -3q+1&=0\\
    -3q&=-1\\
    p &= \frac{1}{3}
\end{align*}
Therefore, the viewer chooses:
\begin{align*}
    p* = 0\ if\ q>\frac{1}{3}\\
    p* \isin (0,1)\ if\ q=\frac{1}{3}\\
    p* = 1\ if\ q<\frac{1}{3}
\end{align*}

Our previous analysis of pure strategies showed that one is not part of a perfect Bayesian equilibrium. Therefore, $q*=\frac{1}{3}$ must be true if this mixed strategy is part of a perfect Bayesian equilibrium. We have now shown that $q*=\frac{1}{3}$ and $p*=\frac{7}{12}$ are the wrestler's and viewer's best response to each other's correct beliefs of what they think each other is going to do and this is a perfect Bayesian Equilibrium. 

\section{Interpret}
The first important thing to notice is that the game could have been simplified so that the wrestler did not have a decision after they saw their opponent was not watching. This is because the payoffs are the same for each the wrestler (and viewer). The wrestler will choose the dominant strategy of showing their strong moves as they get a higher payoff from doing so of 1 and the viewer will of course receive a payoff of 0. This will be further discussed in the next section. 

It made sense that there was no pooling strategy or separating strategy that is a part of a PBE where the confident wrestler doesn't watch the match as watching the match is the dominant strategy for the confident wrestler as they always have something to learn from it even if the wrestler's best moves are not shown. 

It also makes sense that there was no pooling strategy that is a part of a PBE where both types of viewers watch the match. It made more sense for the wrestler to show their best moves as the expected payoff was larger, where they could at least gain some positive payoff when their future competitor is easily intimidated. Knowing this the easily intimidated viewers best response deviated from the strategy as it is their best response to not watch the match knowing the wrestler was going to show off, no matter what the wrestler decides to do when they see their future competitor is not watching the match. 

It also makes sense that there was no separating strategy that is a part of a PBE where the confident viewer watches the match and the easily intimidated one does not watch the match. Knowing that the viewer is confident if they are watching the match, the wrestler would decide to not show their best moves. Seeing this the easily intimidated viewer has incentive to deviate from the strategy and to watch the match as their payoff is higher if they do so. 

The mixed strategy PBE can be stated in the following way. It is the wrestler's best response to show their best moves with a probability of $\frac{1}{3}$ if they correctly believe that an easily intimidated would watch the match with a probability of $\frac{7}{12}$ and that a confident wrestler will always watch the match. Likewise, it is an easily intimidated competitors best response to watch the match with a probability of $\frac{7}{12}$ if they correctly believe that the wrestler will show off with a probability of $\frac{1}{3}$. Finally, of course, it is a confident competitor's best response to watch the match if they correctly believe that the wrestler will show off with a probability of $\frac{1}{3}$. This makes sense as if the wrestler were to showoff with any higher probability, the easily intimidated competitor would never watch the match; if they were to showoff with any lower probability, the easily intimidated competitor would always watch the match. Likewise, if the competitor were to watch the match with a probability any higher the wrestler would always show off; if they were to watch the match with any lower probability the wrestler would never show off. As shown before, all pure strategies were shown not to be part of PBE. 

\section{Enhance}

As discussed, our current game is slightly exhaustive as the wrestler will always show off if their future competitor is not watching, as it is their dominant strategy. The interesting part of the game has ended once the viewer has chosen not to watch the match. Therefore, we will end it there with the reasonable payoffs of 0 for the future competitor and 1 for the wrestler who gets to use their best moves. This is shown in Figure 2 below. 

The probability that a wrestler, of any future competitor is easily intimidated or confident were chosen by an average proportion that I had estimated. In reality though, they may be inaccurate. In addition, the proportions are not a perfect description of every scenario. For example, when my team traveled to states Pennsylvania for tournaments, the wrestling culture there was much different there. Sure, competition was much better, but everyone was 'tough.' The coaches train them at a much more grueling level and a those without a tough mindset, good at the sport or not, were likely to quit. Meanwhile, in Connecticut (where I am from), coaches were encouraged avoid toxic traits associated with sports (I know this as my father was a coach). IN addition, seasons were split in two so that more people could wrestle even if they could not commit as much time; or physical or mental energy. Depending on where one is wrestling, or what teams they are wrestling, there may be a different known probability that the opponent is confident. Therefore, the likely hood that your competitor is confident is situational dependent and can be accurately estimated. Examining the different PBE from these different scenarios is intriguing and informative.

\section{Solve}

It still remains a confident competitors dominant strategy to watch the match: only strategies where confident opponent watches the match will be considered. 

Lets consider if both types of opponents watching the match, a pooling strategy, is a PBE. Once again, in this case, the viewers move of deciding to watch is uninformative to the wrestler. 
\begin{align*}
P(E|W) = P(E) = z\\
P(C|W) = P(C) = 1-z
\end{align*}

Writing down the expected payoff from each move can help determine the wrestlers best response. 

\begin{align*}
    E(\pi_{Wr}(S|W))&=P(E|W)*\pi_{Wr}(E,W,S) + P(C|W)*\pi_{Wr}(C,W,S) \\
    E(\pi_{Wr}(S|W))&=z (3) + 1-z(-2)\\
    E(\pi_{Wr}(S|W))&=5z-2
\end{align*}
\begin{align*}
    E(\pi_{Wr}(N|W))&=P(E|W)*\pi_{Wr}(E,W,N) + P(C|W)*\pi_{Wr}(C,W,N) \\
    E(\pi_{Wr}(N|W))&=z (-1) + (1-z)(-1)\\
    E(\pi_{Wr}(N|W))&=-1
\end{align*}
We will solve for the conditions of $z$, the probability the opponent is easily intimidated, for which not showing off is the wrestlers best response(even weakly). 
\begin{align*}
    E(\pi_{Wr}(S|W))&\leq E(\pi_{Wr}(N|W))\\
    5z-2&\leq-1\\
    5z&\leq1\\
    z&\leq\frac{1}{5}
\end{align*}
As this inequality is linear with $z$, for all other $z$, showing off is the wrestlers best response which makes the easily intimidated opponent deviate from the strategy as knowing this their best response is to not watch the match. However, when $z$, the probability the opponent is easily intimidated, is less than $\frac{1}{5}$, it remains each wrestler's best response to watch the match, as they get a payoff of 1 instead of 0, if they don't watch the match. If $z\leq \frac{1}{5}$, then both wrestlers pooling on watching the match is a PBE. 

Let's reconsider a separating strategy where a confident opponent watches the match and an easily intimidated opponent doesn't watch the match, and see if this is a PBE. In that case, it is the wrestlers best response to not show their best moves if they see their next opponent watching the match as they are certain to be a confident opponent and learn lots from the match. Given this response, the easily intimidated viewers best response to deviate from the strategy and watch the match. 

Though there exists a pooling strategy that is a PBE, it only exists only under a certain condition. Let's see if there exists a mixed strategy that is a PBE. As it is a confident viewer's dominant strategy to always watch the match, lets consider a strategy where they do this and an easily intimidated viewer sometimes watches the match: with a probability $p$. In addition, the wrestler shows their strengths when their opponent is watching with a probability $q$. We will solve to see if there exists a $p$ and $q$ that satisfy a PBE. Therefore;
\begin{align*}
    P(W|E) &= p\\
    P(S|W) &= q\\
    P(W|C) &= 1\\
\end{align*}

Solving for the probability that the viewer is easily intimidated given that they are watching will allow us to create a best response for the wrestler when their opponent is watching. This can be done with Bayes Theorem:
\begin{align*}
    P(E|W)&=\frac{P(W|E)+P(E)}{P(W)}\\
    P(E|W)&=\frac{P(W|E)+P(E)}{P(W|E)P(E)+P(W|C)P(C)}\\
    P(E|W)&=\frac{zp}{zp+(1-z)}\\
    P(E|W)&=\frac{zp}{zp-z+1}
\end{align*}

The probably that the viewer is instead confident is therefore:
\begin{align*}
    P(C|W)&= 1-P(E|W)\\
    P(C|W)&=1-\frac{zp}{zp-z+1}\\
    P(C|W)&=\frac{1-z}{zp-z+1}
\end{align*}
The wrestler wants to show their strengths with a probability that maximizes their expected payoff. 
\begin{align*}
   E(\pi_{Wr,W})= &P(S|W)P(E|W)\pi_{Wr}(E,W,S)\\
                 + &P(S|W)P(C|W)\pi_{Wr}(C,W,S)\\
                 + &P(N|W)P(E|W)\pi_{Wr}(E,W,N)\\
                 + &P(N|W)P(C|W)\pi_{Wr}(C,W,N)\\
    E(\pi_{Wr,W})= &q(\frac{zp}{zp-z+1})(3) \\
                + &q(\frac{1-z}{zp-z+1})(-2) \\
                + &(1-q)(\frac{zp}{zp-z+1})(-1) \\
                + &(1-q)(\frac{1-z}{zp-z+1})(-1)\\
    &\max_q{q(\frac{3zp+2z-2}{zp-z+1}) + (1-q)(-1)}\\
    &\max_q{q\underbrace{(\frac{3zp+2z-2}{zp-z+1}+1)} -1}\\
\end{align*}

    
Solving for when the under-braced part is positive, negative, or zero will tell us what '$q$' the wrestler would choose to maximize their payoff. 
\begin{align*}
    \frac{3zp+2z-2}{zp-z+1}+1&=0\\
    \frac{3zp+2z-2}{zp-z+1}&=-1\\
    3zp+2z-2 &= -zp+z-1\\
    4zp&=1-z\\
    p &= \frac{1-z}{4z}
\end{align*}
Therefore, the wrestler chooses:
\begin{align*}
q* = 0\ if\ p<\frac{1-z}{4z}\\
q* \isin (0,1)\ if\ p=\frac{1-z}{4z}\\
q* = 1\ if\ p>\frac{1-z}{4z}
\end{align*}
Our previous analysis of pure strategies showed that there only exists one for the wrestler if there exists one for the opponent. This happens when $z\leq \frac{1}{5}$. This makes sense because at that point our critical $p$ becomes:
\begin{align*}
    p&=\frac{1-z}{4z}\\
    p&=\frac{1-\leq \frac{1}{5}}{4(\leq \frac{1}{5})}\\
    p&=\frac{\geq \frac{4}{5}}{\leq \frac{4}{5}}\\
    p&\geq 1\\
    p&=1
\end{align*}

This mixes strategy where $p*=\frac{1-z}{4z}$ does therefore include our pure strategy PBE under the correct conditions. It also makes sense as if we substitute $z$ for $0.3$, as we assumed before, we get the same $p*$ of $\frac{7}{12}$. 

A best response should also be determined for the viewer that is easily intimidated. The viewer wants to watch the match with a probability that maximizes their expected payoff. 
\begin{align*}
   E(\pi_{V,E})= &P(S|W)P(W|E)\pi_{V}(E,W,S)\\
                 + &P(NW|E)\pi_{V}(E,NW)\\
                 + &P(N|W)P(W|E)\pi_{V}(E,W,N)\\
    E(\pi_{V,E})= &qp(-2) \\
                + &(1-p)(0) \\
                + &(1-q)p(1) \\
    &\max_q{p(-2q) + p -pq}\\
    &\max_q{p\underbrace{(-3q+1)}}\\
\end{align*}
Solving for when the under-braced part is positive, negative, or zero will tell us what '$p$' the viewer would choose to maximize their payoff. 
\begin{align*}
    -3q+1&=0\\
    -3q&=-1\\
    p &= \frac{1}{3}
\end{align*}
Therefore, the viewer chooses:
\begin{align*}
    p* = 0\ if\ q>\frac{1}{3}\\
    p* \isin (0,1)\ if\ q=\frac{1}{3}\\
    p* = 1\ if\ q<\frac{1}{3}
\end{align*}

As shown before, our mixed strategy of $q*=\frac{1}{3}$ includes our PBE under the right conditions that $z\leq \frac{1}{5}$. We have now shown that, as long as $z\leq \frac{1}{5}$, $q*=\frac{1}{3}$ and $p*=\frac{1-z}{4z}$ are the wrestler's and viewer's best response to each other's correct beliefs of what they think each other is going to do and this mixed strategy is a perfect Bayesian Equilibrium. 

\section{Interpret}
It makes sense that there is only a pooling strategy PBE when $z$ is sufficiently small ($\leq \frac{1}{5}$). This is because there must be enough confident opponents and not as many easily intimidated opponents for the wrestler's best response to be not to show off if all their opponents are going to always watch their match. In this strategy, the opponent earns a payoff of 1 from gaining some knowledge of the wrestlers basic technique, which the wrestler loses as their opponent now has a slight edge. This makes sense as when I was at wrestling tournaments in Pennsylvania, where a much larger proportion of my opponents are tough, I was much less likely to showoff or show my best moves in early matches as they were my edge and I knew it was unlikely I would intimidate any opponents. Likewise these opponents were tough and often watched my matches to learn from them. However, if more than one fifth of opponents are easily intimidated, it becomes the wrestlers best response to always show off and therefore the easily intimidated opponents best response would instead be to not watch the match. 

It also makes sense that there was no separating strategy that is a part of a PBE where the confident viewer watches the match and the easily intimidated one does not watch the match. Knowing that the viewer is confident if they are watching the match, the wrestler would decide to not show their best moves. Seeing this the easily intimidated viewer has incentive to deviate from the strategy and to watch the match as their payoff is higher if they do so. 

Let's state the mixed strategy solution in a clearer way one again. It is the wrestler's best response to show their best moves with a probability of $\frac{1}{3}$ if they correctly believe that an easily intimidated would watch the match with a probability of $\frac{1-z}{4z}$, where $z$ is the probability that the opponent is easily intimidated, and that a confident wrestler will always watch the match. Likewise, it is an easily intimidated competitors best response to watch the match with a probability of $\frac{1-z}{4z}$ if they correctly believe that the wrestler will show off with a probability of $\frac{1}{3}$. Finally, of course, it is a confident competitor's best response to watch the match if they correctly believe that the wrestler will show off with a probability of $\frac{1}{3}$. These are are all under the condition that $z \geq \frac{1}{5}$ as otherwise these probabilities are not between 0 and 1 and the pooling strategy stated above is the PBE. 

It makes sense that the probability the wrestler must show off for this mixed strategy PBE does not change from $\frac{1}{3}$ when we generalized the probability the opponent is easily intimidated. The wrestler must show off with a probability of one third as opponent knows if they are confident or not and their expected payoffs are the same, this leaves the easily intimidated opponent indifferent with what probability they choose to watch the match. 

We can know see the probability an opponent is easily intimidated affects the probability that an easily intimidated opponent will watch the match in the mixed strategy PBE. 

\begin{align*}
    p* &= \frac{1-z}{4z}\\
    \frac{\partial p*}{\partial z} &= \frac{4z(-1)-(1-z)(4)}{16z^2}\\
    \frac{\partial p*}{\partial z} &= \frac{-1}{16z^2}\\
\end{align*}

This partial derivative is negative: meaning that the higher probability a wrestler is easily intimidated, the lower the probability that easily intimidated opponent will watch the match in the mixed strategy PBE. This makes sense as when the probability an opponent is easily intimidated increases, the less likely they must be to watch the match for the wrestler to remain indifferent with what probability they choose to showoff. 

The expected payoff for the wrestler is derived below. Recall $q*=\frac{1}{3}$ and $p* = \frac{1-z}{4z}$. 
\begin{align*}
   E(\pi_{Wr})= &P(S|W)P(E|W)\pi_{Wr}(E,W,S)\\
                 + &P(S|W)P(C|W)\pi_{Wr}(C,W,S)\\
                 + &P(N|W)P(E|W)\pi_{Wr}(E,W,N)\\
                 + &P(N|W)P(C|W)\pi_{Wr}(C,W,N)\\
                 + &P(NW|E)\pi_{Wr}(C,NW)\\
    E(\pi_{Wr})= &q(\frac{zp}{zp-z+1})(3) \\
                + &q(\frac{1-z}{zp-z+1})(-2) \\
                + &(1-q)(\frac{zp}{zp-z+1})(-1) \\
                + &(1-q)(\frac{1-z}{zp-z+1})(-1)\\
                + (1-p)(1)
     E(\pi_{Wr})= &\frac{3qzp-2p(1-z)}{zp+1-z} + q-1\\
     E(\pi_{Wr})= &\frac{\frac{1}{4}(1-z)-\frac{2}{3}(1-z)}{\frac{1}{4}(1-z)+(1-z)} - \frac{2}{3} + 1 - \frac{1-z}{4z}\\
     E(\pi_{Wr})= &\frac{\frac{1}{4}-\frac{2}{3}}{\frac{1}{4}+1} - \frac{2}{3} + 1 - \frac{1-z}{4z}\\
     E(\pi_{Wr})= &\frac{\frac{3}{12}-\frac{8}{12}}{\frac{5}{4}} - \frac{2}{3} + 1 - \frac{1-z}{4z}\\
     E(\pi_{Wr})= &\frac{\frac{-5}{12}}{\frac{5}{4}} - \frac{2}{3} + 1 - \frac{1-z}{4z}\\
     E(\pi_{Wr})= &\frac{-20}{60} - \frac{2}{3} + 1 - \frac{1-z}{4z}\\
     E(\pi_{Wr})= &\frac{-20}{60} - \frac{2}{3} + 1 - \frac{1-z}{4z}\\
     E(\pi_{Wr})= &- \frac{1-z}{4z}
\end{align*}

As $z\isin (-\frac{1}{5}$, $E(\pi_{Wr}) \isin (-1,0)$. Lets how the profit of the wrestler changes as the probability their opponent is easily intimidated. 
\begin{align*}
    \frac{\partial E(\pi_{Wr})}{\partial z} &= \frac{4z(1)+(1-z)(4)}{16z^2}\\
    \frac{\partial E(\pi_{Wr})}{\partial z} &= \frac{1}{16z^2}\\
\end{align*}

This partial derivative is positive meaning that as the probability an opponent is easily intimidated increases, the expected profit of the wrestler increases. This makes sense as increasing the probability an opponent is easily intimidated means that opponent must be must be more likely to not watch the match in the mixed strategy PBE, increasing the payoff for the wrestler. 


Lets now derive the expected payoff for an easily intimidated opponent. 
\begin{align*}
   E(\pi_{V,E})= &P(S|W)P(W|E)\pi_{V}(E,W,S)\\
                 + &P(NW|E)\pi_{V}(E,NW)\\
                 + &P(N|W)P(W|E)\pi_{V}(E,W,N)\\
    E(\pi_{V,E})= &qp(-2) \\
                + &(1-p)(0) \\
                + &(1-q)p(1) \\
    E(\pi_{V,E})= &p(-2q) + p -pq\\
    E(\pi_{V,E})= &-3q+1\\
    E(\pi_{V,E})= &0\\
\end{align*}
It makes sense that the expected profit of the easily intimidated wrestler is 0 in the mixed strategy PBE as for them to be indifferent about what probability they choose to watch the match their expected profit from watching the match. must be the same as their expected profit from not watching the match and the expected profit from not watching the match is always 0. 

Let's now derive the expected payoff for a confident opponent. 

\begin{align*}
   E(\pi_{V,C})= &P(S|W)\pi_{V}(C,W,S)\\
                 + &P(N|W)\pi_{V}(C,W,N)\\
    E(\pi_{V,C})= &q(3) \\
                + &(1-q)(1) \\
    E(\pi_{V,C})= &3q + 1 -q\\
    E(\pi_{V,C})= &2q+1\\
    E(\pi_{V,C})= &\frac{2}{3} + 1\\
    E(\pi_{V,C})= &\frac{5}{3}\\
\end{align*}

It makes sense that the expected payoff of the confident opponent does not change with the probability an opponent is easily intimidated as for the easily intimidated opponent to be indifferent about what probability they watch the match, a requirement for the mixed strategy PBE, the wrestler must always show off with a probability of one third, meaning the always makes the same expected payoff. 


 


\section{Conjecture}
While the model is not a perfect representation of reality, it does show logical equilibrium that are close to the reality of what I observed and the decisions I made while wrestling. I really did change my decision to showoff based on where I was. When I showed my strengths and then went out strong and it would be clear that sometimes my opponent had gotten in their head, mostly by the decrease in strength they put in their moves. Sometimes it was also clear that I did not intimidate anyone any my intensity was matched. As well, in these cases the wrestler was clearly changing their form to make it more difficult for me to use my favorite moves (taller stance, leaning away from my riding side, and stalling when I was near my strongest positions). As said this model could also be generalized to other sports, though likely not team based ones as the mentality is not as homogeneous and easily intimidated. 

This model could be made better in multiple ways. Confidence is based off genetics that are not so black and white. We could have made confidence come from a continuous distribution and therefore made our payoffs continuous as well to make this model more realistic. In this case, if we wanted to change the relative probabilities of each opponent being a certain confidence based on the perceived toughness of the area, we could change the shape and mean of our confidence parameter distribution. While it makes sense for showing your strengths to be discretized, there could've been varying amounts of this as you might have more than one move that is your strength. Lastly, it could've make sense to add some form of risk intolerance to the model. While it makes sense for people to want to maximize their expected payoff, they also dislike, in some amount, risk. For example, not losing horrible or getting pinned means more team points and some people simply dislike the risk of doing really bad and would rather lose, by a minimal amount, that have the opportunity to lose by a large amount. These types of people might dislike the risk of showing their strengths to their opponent even if it is likely they could intimidate their opponent. Some may even dislike the opportunity to do better- and would rather accept and deal with the loss currently than hope. This brings many questions about the human implementation of risk intolerance. Do humans wish to minimize the worst scenario? Or would they rather lessen the expected difference between the best and worst case scenario? Do they care more about large deviations from the expected out come? If so by how much? Do they do all of these things but with different priorities? This stuff intersection of game theory, psychology, and econometric is very interesting and using a model like this for our tests/data could tell us lots of about human nature, hope, and risk tolerance.
