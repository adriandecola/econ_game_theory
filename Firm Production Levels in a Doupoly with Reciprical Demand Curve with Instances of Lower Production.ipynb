# Firm Production Levels in a Doupoly with Reciprical Demand Curve with Instances of Lower Production
## Motivation
Consider two firms, one producing Pepsi and the other producing Coca-Cola. They each produce the same product, a vanilla-and-cinnomon-like drink, indifferentiable by our tastebuds (despite what some fanatics for either argue). We can imagine the demand curve for these products is non-linear. How much will these firms produce given their thoughts of the other firms? What is/are the Nash Equilibrium(s)? 

## Build
A non-linear demand funciton, specifically a reciprical one, aligns more closely with theory as the increased benefit of one more soda (marginal utility) depends on how many sodas they've had, more specifically it decreases with more sodas. One would want a second soda more than they would want a twentyth soda. A linear demand curve does not capture this fact, but a reciprical one does. Therefore, the assumed demand function is: $$p(q_i, q_{-i})=\frac{\xi}{q_i+q_{-i}}$$Where $\xi$ is a factor related to the price elasticity of demand and the substript $_i$ denotes a one of the firms and the subscript $_{-i}$ denotes the other firm. It is also clear from this price function that the produciton of (0, 0) is indeterminante and will not be considered in this problem. In addition, we also assume that each firm can produce at whatever quantity and that the marginal cost of producing one more good, or soda, is constant. I say this because we can consider the cost of producing another factory to be included in the variable cost. For Pepsi and Coca-Cola, this also makes sense as many factories can be produced and basic goods are used in the production of soda bottles so that there is essentially no scarcity. Therefore the total variable cost is $q_iv$ where $v$ is the cost of producing one more good. A fixed cost labeled $FC$ is assumed for operation costs present no matter the production levels. We assume that each firm has the same variable cost and fixed cost since they are producing essentially the same product. 

## Solve
Based on the assumptions above, we can define a profit funciton for each firm:
$$\pi_i = q_ip(q_i, q_{-i}) - q_iv-FC$$
Where the first term is the revenue for the firm, the second term is the total variable cost, and the thrid term is clearly the fixed cost. Each firm wishes to maximize their profit based on their assumption of the other firms production $\hat{q}_{-i}$. Taking the first order conditions we get:
\begin{align}
\max_{q_i}&\ \pi_i\\
\max_{q_i}&\ q_ip(q_i, \hat{q}_{-i}) - q_iv-FC\\
0&=\frac{\partial }{\partial q_i}\left(q_ip(q_i, \hat{q}_{-i}) - q_iv-FC\right)\\
0&= p(q_i, \hat{q}_{-i}) + \frac{\partial p(q_i, \hat{q}_{-i})}{\partial q_i} q_i -v\\
0&= \frac{\xi}{q_i+\hat{q}_{-i}} - \frac{\xi}{(q_i+\hat{q}_{-i})^2}q_i-v\\
0&= \xi(q_i + \hat{q}_{-i}) - \xi q_i -v(q_i+\hat{q}_{-i})^2\\
0&= \frac{\xi \hat{q}_{-i}}{-v} + q_i^2 + 2\hat{q}_{-i}q_i + \hat{q}_{-i}^2\\
0&= q_i^2 + 2\hat{q}_{-i}q_i + \left( \hat{q}_{-i}^2 - \frac{\xi \hat{q}_{-i}}{v}\right)
\end{align}
Using the quadratic formula we get:
$$q_i = -\hat{q}_{-i} \pm \frac{1}{2} \sqrt{4q_{-i}^2-4q_{-i}^2+\frac{4\xi \hat{q}_{-i}}{v}}$$
$$q_i = -\hat{q}_{-i} \pm \frac{1}{2} \sqrt{\frac{4\xi \hat{q}_{-i}}{v}}$$
$$q_i = -\hat{q}_{-i} \pm  \sqrt{\frac{\xi \hat{q}_{-i}}{v}}$$
Since each production must be positive and nonzero, it is clear that the only solution is:
$$q_i = -\hat{q}_{-i} +  \sqrt{\frac{\xi \hat{q}_{-i}}{v}}$$
These is the best response equation for each firm assuming the other firm produces at a level of $\hat{q}_{-i}$. Under the Nash equilibrium each firm correctly guessed the other firms production and produces at their best response. Therefore, $\hat{q}_{-i} = \hat{q}_{-i}^*$ and $\hat{q}_{i} = \hat{q}_{i}^*$:
\begin{align}
q_i^* &= -\left( -q_{i}^* +  \sqrt{\frac{\xi q_{i}^*}{v}}\right) + \sqrt{\frac{\xi \left( -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{v}}\right)}{v}}\\
q_i^* &= q_{i}^* - \sqrt{\frac{\xi q_{i}^*}{v}} + \sqrt{\frac{\xi \left( -q_{i}^* +\sqrt{\frac{\xi q_{i}^*}{v}}\right)}{v}}\\
\sqrt{\frac{\xi q_{i}^*}{v}} &= \sqrt{\frac{\xi \left( -q_{i}^* +  \sqrt{\frac{\xi q_{i}^*}{v}}\right)}{v}}\\
\frac{\xi q_{i}^*}{v} &= \frac{\xi \left(-q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{v}}\right)}{v}\\
q_{i}^* &= -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{v}}\\
2q_{i}^* &= \sqrt{\frac{\xi q_{i}^*}{v}}\\
0&= \left(4q_{i}^*\right)^2- \frac{\xi q_{i}^*}{v}\\
0&=q_{i}^*(4q_{i}^*-\frac{\xi}{v})\\
q_{i}^*&=\frac{\xi}{4v}
\end{align}

As discussed in the Assumptions, production levels of 0 will not be considered and therefore, the Nash equilibrium is when each firm produces $\frac{\xi}{4v}$. 

## Interpret
The general best response function derived above is $q_i = -\hat{q}_{-i} + \sqrt{\frac{\xi \hat{q}_{-i}}{v}}$. We can see how this changes with $\hat{q}_{-i}$, $\xi$, and $v$. 
$$\frac{\partial }{\partial \hat{q}_{-i}} q_i = -1+\frac{1}{2} \sqrt{\frac{\xi}{\hat{q}_{-i}v}}$$
$$\frac{\partial }{\partial \xi} q_i = \frac{1}{2} \sqrt{\frac{\hat{q}_{-i}}{\xi v}}$$
$$\frac{\partial }{\partial v} q_i = -\frac{1}{2} \sqrt{\frac{\xi \hat{q}_{-i}}{v^3}}$$
It is clear from the first partial derivitive that the best response output, $q_i$, increases until $\hat{q}_{-i}=\frac{\xi}{4v}$, the Nash Equilibrium, at which point the best response output decreases when the firm thinks the other firm will produce more as the partial derivitive becomes negative. This is an interesting result that makes sense. If one firm thinks the other firm will produce more while still producing below the Nash equilibrium production, it then becomes important to produce more as the price will have significantly decreased. Meanwhile, if one firm thinks the other firm will produce more while already expecting them to produce above the Nash equilibrium quantity, it then mbecomes more important to raise the price of the soda being sold and it is in the firms best interest to produce less soda. 

As a square root is always positve, when $\xi$ increases the best response output, $q_i$, increases. This makes sense becauase a larger $\xi$ means the price is higher for each combination of outputs quantities and the best response should be higher to capture more revenue by producing more soda. 

When $v$ increases, the best response output, $q_i$, decreases. This makes sense because as the variable cost increases you will no longer want to produce as much soda as the total variable cost has increased. 

The Nash Equilibrium says that if each firm correctly believes that the other will produce at a level of $\frac{\xi}{4v}$, then the firms best response it to also produce at a level of $\frac{\xi}{4v}$. Likewise, if $\xi$ increases, we would expect each of the firms to procude more in the Nash equilibrium as there is a higher price and they can capture more revanue by producing more. In addition, if the variable cost, $v$, increases, we would expect each firm to produce less in th Nash equilibrium as the total variable cost has increased for any given output level. 

## Enhance
Let's assume these firms are producing at the Nash equilibrium. Let's imagine each firm has days (or weeks or months) with lower production. Depending on the time period there would be reasons for each firm to have variations in production. One day might have a lower production because the shippping facility is experiencing a problem. A week might have lower production due to holidays. A month might have lower production due to supply shortages. Of course if this was a pure strategy we would find production levels according to the Nash Equilibrium; however, it is more realistic for the companies to have time periods with variations in production. Let's assume these variations in production exist. Let's also assume that firms can decide the probability they produce at a lower produciton, by hiring a certain amount of staff, lower than the optimal, that rotates shifts, or by other means. Let's also assume that the amount of production for each firm when they produce at a lower level is the same for each firm. Finally, let's assume that these days of lower production are truly probablistic for each firm (this wouldn't make sense in the holiday senerio). 

We can define the amount the company produces in the day of lower production to be a factor $d$ of the equilibrium prduction, where $d\in[0,1]$. Let's also call the probabilty the firm lowers production is $P_{i,l}$ and the probability the other firm lowers production as $P_{-i,l}$. We can call profit for firm i when they each lower their output $\pi_{l,l}$. Likewise when each firms produce the equilibrium quanitity, firm i's profit is $\pi_{e,e}$. When firm i produces a lower output but the other firm produces the equilibrium quantity, firm i's profit is $\pi_{l,e}$. When firm i produces at the equilibrium quantity and the other firm produces at the lower quantity, firm i's profit is $\pi_{e,l}$.

## Solve 
The firms problem is therefore, to maximize the expected profit by choosing the probability they produce at a lower output. 
\begin{align}
\max_{P_{i,l}}&\ E(\pi)\\
\max_{P_{i,l}}&\ P_{i,l}(P_{-i,l}\pi_{l,l} + (1-P_{-i,l})\pi_{l,e}) (1-P_{i,l})P_{-i,l}\pi_{e,l} + (1-P_{-i,l})\pi_{e,e}\\
\max_{P_{i,l}}&\ P_{i,l}\underline{\left(P_{-i,l}\pi_{l,l} + (1-P_{-i,l})\pi_{l,e}-P_{-i,l}\pi_{e,l}-(1-P_{-i,l})\pi_{e,e}\right)} + \left(P_{-i,l}\pi_{e,l} + (1-P_{-i,l})\pi_{e,e}\right)\\
\end{align}
        Note: Sometimes this underlines correctly and sometimes it doesn't-- I can't figure out why as it is the same expression. Also, sometimes the hats act super weirdly and sometimes they don't. 
        
        
Since the firm can only choose the probability they lower production, $P_{i,l}$, we are interested in when the underlined part is negative, as in that case the expected profit is maximized when that probability is 0, or the firm never lowers production, when it is zero as the firm is indifferent to the probability it lowers production, and when it is positive where expected profit is maximized when the probability is 1, or the firm always lowers production. 

We know from the last problem that the equilibrium production is $\frac{\xi}{4v}$. As definied the lower output is $d\frac{\xi}{4v}$. Finding the profits in each senerio:
\begin{align}
\pi_{e,e}&= \frac{\xi}{4v} \left(\frac{1}{\frac{\xi}{4v}+\frac{\xi}{4v}}-v\right)-FC\\
\pi_{e,e}&=\frac{\xi}{4}-FC\\
\pi_{e,l}&= \frac{\xi}{4v} \left(\frac{1}{\frac{\xi}{4v}+\frac{d\xi}{4v}}-v\right)-FC\\
\pi_{e,l}&=\frac{\xi}{1+d}-\frac{\xi}{4}-FC\\
\pi_{l,e}&= \frac{d\xi}{4v} \left(\frac{1}{\frac{d\xi}{4v}+\frac{\xi}{4v}}-v\right)-FC\\
\pi_{l,e}&=\frac{d\xi}{1+d}-\frac{d\xi}{4}-FC\\
\pi_{l,l}&= \frac{d\xi}{4v} \left(\frac{1}{\frac{d\xi}{4v}+\frac{d\xi}{4v}}-v\right)-FC\\
\pi_{l,l}&=\frac{\xi}{2}-\frac{d\xi}{4}-FC
\end{align}
Setting the underlined expression equal to zero:
\begin{align}
0&=P_{-i,l}(\frac{\xi}{2}-\frac{d\xi}{4}-FC) + (1-P_{-i,l})(\frac{d\xi}{1+d}-\frac{d\xi}{4}-FC)-P_{-i,l}(\frac{\xi}{1+d}-\frac{\xi}{4}-FC)-(1-P_{-i,l})(\frac{\xi}{4}-FC)\\
0&=P_{-i,l}(\frac{1}{2}-\frac{d}{4}) + (1-P_{-i,l})(\frac{d}{1+d}-\frac{d}{4})-P_{-i,l}(\frac{1}{1+d}-\frac{1}{4})-(1-P_{-i,l})(\frac{1}{4})\\
0&=P_{-i,l}(-\frac{d}{1+d}-\frac{1}{1+d}) + \frac{d}{1+d} - \frac{d}{4}-\frac{1}{4}\\
P_{-i,l}&=\frac{\frac{d}{1+d} - \frac{d}{4}-\frac{1}{4}}{\frac{d}{1+d}+\frac{1}{1+d}}\\
P_{-i,l}&=\frac{d}{1+d} - \frac{d}{4}-\frac{1}{4}\\
P_{-i,l}&=\frac{4d-d^2-1-d-1}{4(d+1)}\\
P_{-i,l}&=\frac{-d^2+3d-2}{4(d+1)}\\
P_{-i,l}&=\frac{-(d-1)(d-2)}{4(d+1)}\\
\end{align}

These probabilities must be between 0 and 1. Since $d\in[0,1]$, $(d-2)$ is always negative, $(d+1)$ is always positive, and $(d-1)$ is always negative except when $d=1$. Therefore $P_{-i,l}$ is always negative except when $d=1$, where it is zero. However when $d=1$, means there is not actually any lower production, the firm continues to produce at the equilibrium level. Therefore there is no mixed strategy solution as we will discuss below. 

The underlined section, is always negative because as shown above $\frac{d}{1+d} - \frac{d}{4}-\frac{1}{4}=\frac{-(d-1)(d-2)}{4(d+1)}$ is always negative and all the other terms are negative. Therefore the expected payoff is maximized when $P_{i,l} = 0$. 

## Interpret 
When $d=1$ there was no lower production, therefore it did not make sense to define an indifference point and mixed strategy equilibrium as each strategy was the same level of production. It makes sense that there was no mixed strategy equilibrium because when one firm produces less soda than in the equilibrium, the other firm wants to produce more soda to capture more potential profit due to the lower price. The firm would never want to produce less that the equilibrium quantity as that would result in an even lower profit. Therefore there is only a pure strategy Nash equilibrium in this game as derived above.
