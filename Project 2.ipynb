# Doupolistic Production Levels with varying Variable Costs and Risk Intolerance
## Motivation
   Having detailed, realistic plans is important for firms. Consider two firms producing a similar product with near complete market share, like Boeing and Airbus. These doupolistic firms make choices to produce commercial airplanes at certain levels, through contracts and other forms, like hiring and scheduling worker hours, that are preplanned. However, these firms do not know the exact conditions under which they will be in manufacturing the planes. Weather or geopolitical conditions could drastically change the price of an input good in plane manufacturing, the companies could secure government subsidies, their workers could even go on strike. Due to this, when Airbus or Boeing make production level decisions and enter contracts, they have an idea of the variable cost of their planes but are not certain of what this cost will be and must plan accordingly. 
   
   These uncertainties mean firms might be left with more or less profit than expected. Firms must deal with this uncertainty in their profit. Clearly making negative profit in one period then surplus profit in the next is worse than making the same average profit each period as the turmoil from making negative profit from one period will carry onto the next period, such as: requiring the company to make decisions, like taking out loans, firing workers, or cutting projects, that inccur large costs. Some firms may be more prepared to make these sorts of decisions with minimal "piling-on" negative effects in the future. Examples of this might be the ability to cancel bonuses, rather than having to fire employees or the having a good loan history allowing them to take a low interest loan. 
    
   Do duopolistic firms, like Airbus and Boeing, change their production levels with uncertainty in their production costs? Does a firm with less uncertainty in their production cost produce more the same as their counter part? How about when they are intolerant for risk? Do firms with more risk tolerance tend to produce more or less than their risk intolerant counterpart? Do they tend to earn more, less or the same as their risk intolerant counter part? What are the Nash Equilibrium production levels when we account for uncertainties and risk intolerance?

## Build
First lets consider and duopolistic situation where the two firms have different variable costs, with different uncertainties, and no risk intolerance. The profit for a firm, $i$, is of the form $$\pi_i = p(q_i, q_{-i}) q_i - v_i q_i - FC$$Where the the substript $_i$ denotes a one of the firms and the subscript $_{-i}$ denotes the other firm. $p(q_i, q_{-i})$ price given the production levels of each firm. To capture diminishing marginal utility we do not use a linear demand curve but rather one of the form $$p(q_i, q_{-i})=\frac{\xi}{q_i+q_{-i}}$$Where $\xi$ is a factor related to the price elasticity of demand. It is also clear from this price function that the produciton of zero from each firm will not be considered in this problem as it implies an infinite price. $FC$ is the fixed cost: lets assume it is the same for both firms. As described, the variable cost is not certain. Lets assume they face the same average variable cost as they are producing essentially the same product, but different uncertainties in these variable costs. The firms' knowledge of what the variable cost will be is of the form of a normal distrobution: $$v_i \sim N(\overline{v}, \sigma_{v_i}^2)$$Where $\overline{v}$ is the mean of the normal distribution and $\sigma_{v_i}^2$ is the variance in this value. Let's call $\sigma_{v_i}$, the standard deviation in the variable cost of firm $i$, the uncertainty in the variable cost. 
    
   The goal of each firm is therefore to maximize their expected profit. $$\max_{q_i}\ E(\pi_i)= E(p(q_i, q_{-i})) q_i - v_i q_i - FC)$$

## Solve
\begin{align}
\max_{q_i}&\ E(\pi_i)= E(p(q_i, q_{-i})) q_i - v_i q_i - FC)
\end{align}
Let's define $\hat{q}_{-i}$ what production level firm $i$ thinks firm $-i$ will produce at and $\hat{q}_{i}$ what production level firm $-i$ thinks firm $i$ will produce at.
\begin{align}
\max_{q_i}&\  p(q_i, \hat{q}_{-i}) q_i - E(v_i) q_i - FC\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - E(v_i) q_i - FC\\
\end{align}
Taking the first order conditions we get:
\begin{align}
0&=\frac{\partial }{\partial q_i}\left(q_i\frac{\xi}{q_i+\hat{q}_{-i}} - q_iE(v_i)-FC\right)\\
0&= p(q_i, \hat{q}_{-i}) + \frac{\partial \frac{\xi}{q_i+\hat{q}_{-i}}}{\partial q_i} q_i -E(v_i)\\
0&= \frac{\xi}{q_i+\hat{q}_{-i}} - \frac{\xi}{(q_i+\hat{q}_{-i})^2}q_i-E(v_i)\\
0&= \xi(q_i + \hat{q}_{-i}) - \xi q_i -E(v_i)(q_i+\hat{q}_{-i})^2\\
0&= \frac{\xi \hat{q}_{-i}}{-E(v_i)} + q_i^2 + 2\hat{q}_{-i}q_i + \hat{q}_{-i}^2\\
0&= q_i^2 + 2\hat{q}_{-i}q_i + \left( \hat{q}_{-i}^2 - \frac{\xi \hat{q}_{-i}}{E(v_i)}\right)
\end{align}
As the expected value of the variable cost is its mean $$E(v_i)=\overline{v}$$
Using the quadratic formula we get:
$$q_i = -\hat{q}_{-i} \pm \frac{1}{2} \sqrt{4q_{-i}^2-4q_{-i}^2+\frac{4\xi \hat{q}_{-i}}{\overline{v}}}$$
$$q_i = -\hat{q}_{-i} \pm \frac{1}{2} \sqrt{\frac{4\xi \hat{q}_{-i}}{\overline{v}}}$$
$$q_i = -\hat{q}_{-i} \pm  \sqrt{\frac{\xi \hat{q}_{-i}}{\overline{v}}}$$
Since each production must be positive and nonzero, it is clear that the only solution is:
$$q_i = -\hat{q}_{-i} +  \sqrt{\frac{\xi \hat{q}_{-i}}{\overline{v}}}$$
This is the best response equation for each firm assuming the other firm produces at a level of $\hat{q}_{-i}$. Under the Nash equilibrium each firm correctly guessed the other firms production and produces at their best response. Therefore, $\hat{q}_{-i} = \hat{q}_{-i}^*$ and $\hat{q}_{i} = \hat{q}_{i}^*$:
\begin{align}
q_i^* &= -\left( -q_{i}^* +  \sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\right) + \sqrt{\frac{\xi \left( -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\right)}{\overline{v}}}\\
q_i^* &= q_{i}^* - \sqrt{\frac{\xi q_{i}^*}{\overline{v}}} + \sqrt{\frac{\xi \left( -q_{i}^* +\sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\right)}{\overline{v}}}\\
\sqrt{\frac{\xi q_{i}^*}{\overline{v}}} &= \sqrt{\frac{\xi \left( -q_{i}^* +  \sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\right)}{\overline{v}}}\\
\frac{\xi q_{i}^*}{\overline{v}} &= \frac{\xi \left(-q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\right)}{\overline{v}}\\
q_{i}^* &= -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\\
2q_{i}^* &= \sqrt{\frac{\xi q_{i}^*}{\overline{v}}}\\
0&= \left(4q_{i}^*\right)^2- \frac{\xi q_{i}^*}{\overline{v}}\\
0&=q_{i}^*(\underbrace{4q_{i}^*-\frac{\xi}{\overline{v}}}_{=0})\\
q_{i}^*&=\frac{\xi}{4\overline{v}}
\end{align}

As explained in the assumptions, production levels of 0 imply an infinite price and will not be considered, therefore the Nash equilibrium is when each firm produces $\frac{\xi}{4\overline{v}}$. 

## Interpret
It becomes clear that this problem reduces to solving the best response functions and Nash Equilibrium using the mean variable cost where the uncertainty in the variable cost does not matter. This makes sense as the firms problem is to maximize the expected value of their profit, which is linear in the variable cost and since the expected value of a normal distribution is its mean, the problem simplified to maximizing the profit with a variable cost of the mean of the its distribution. 
The general best response function derived above is $q_i = -\hat{q}_{-i} + \sqrt{\frac{\xi \hat{q}_{-i}}{\overline{v}}}$. We can see how this changes with $\hat{q}_{-i}$, $\xi$, and $v$. 
$$\frac{\partial }{\partial \hat{q}_{-i}} q_i = -1+\frac{1}{2} \sqrt{\frac{\xi}{\hat{q}_{-i}\overline{v}}}$$
$$\frac{\partial }{\partial \xi} q_i = \frac{1}{2} \sqrt{\frac{\hat{q}_{-i}}{\xi \overline{v}}}$$
$$\frac{\partial }{\partial \overline{v}} q_i = -\frac{1}{2} \sqrt{\frac{\xi \hat{q}_{-i}}{\overline{v}^3}}$$

As a square root is always positve, when $\xi$ increases the best response output, $q_i$, increases. This makes sense becauase a larger $\xi$ means the price is higher for each combination of outputs quantities and the best response should be higher to capture more revenue by producing more commercial airplanes.

When $\overline{v}$ increases, the best response output, $q_i$, decreases. This makes sense because as mean variable cost increases the expected profit decreases when you produce the same and the firm could capture more profit by producing less at a higher price. 

It is clear from the first partial derivitive that the best response output, $q_i$, increases until $\hat{q}_{-i}=\frac{\xi}{4\overline{v}}$, the Nash Equilibrium, at which point the best response output decreases when the firm thinks the other firm will produce more as the partial derivitive becomes negative. This is an interesting result that makes sense. If one firm thinks the other firm will produce more while still producing below the Nash equilibrium production, it then becomes important to produce more as the price will have significantly decreased. Meanwhile, if one firm thinks the other firm will produce more while already expecting them to produce above the Nash equilibrium quantity, it then becomes more important to raise the price of the soda being sold and it is in the firms best interest to produce less soda to capture more revenue. 

The Nash Equilibrium says that if each firm correctly believes that the other will produce at a level of $\frac{\xi}{4\overline{v}}$, then the firms best response it to also produce at a level of $\frac{\xi}{4\overline{v}}$. Likewise, if $\xi$ increases, we would expect each of the firms to procude more in the Nash equilibrium as there is a higher price and they can capture more revanue by producing more. In addition, if the average variable cost, $\overline{v}$, increases, we would expect each firm to produce less in th Nash equilibrium as the expected total variable cost has increased for any given output level. 

It is also clear that it does not matter how the variable cost is distributed or its spread. All that mattered was its expected value or mean. This once again comes from the fact that the profit is linear in the variable cost. It's also because the firms goal was to simply maximize the expected profit. We will relax this assumption in the next section.  

## Enhance
Clearly a game similar to the one described is repeated throughout time for a duopoly like Airbus and Boeing. These firms would not be risk intolerant and would like to minimize the uncertainty in the profit they get, as described above in Motivation. We will, similar to variable cost, call the uncertainty in the profit its standard devitation. Therefore the firms maximization problem becomes $$\max_{q_i}\ E(\pi_i)-\delta_istd(\pi_i)$$Where $\delta_i$ is a factor related to a firm $i$'s tolerance for uncertainty in their profit, and $std(\pi_i)$ is the standard deviation, or uncertainty in firm $i$'s profit. The higher $\delta_i$ is the more intolerant a firm is to uncertainty in their profit. 

## Solve 
Let's define $\hat{q}_{-i}$ what production level firm $i$ thinks firm $-i$ will produce at and $\hat{q}_{i}$ what production level firm $-i$ thinks firm $i$ will produce at.
\begin{align}
\max_{q_i}&\  p(q_i, \hat{q}_{-i}) q_i - E(v_i) q_i - FC-\delta_i\sqrt{var(\pi_i)}\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - \overline{v} q_i - FC-\delta_i\sqrt{var(\pi_i)}\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - \overline{v} q_i - FC-\delta_i\sqrt{var(\frac{\xi}{q_i+\hat{q}_{-i}} q_i - v q_i - FC)}\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - \overline{v} q_i - FC-\delta_i\sqrt{var(v q_i)}\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - \overline{v} q_i - FC-\delta_i\sqrt{q_i^2var(v)}\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - \overline{v} q_i - FC-\delta_i\sqrt{q_i^2\sigma_{v_i}^2}\\
\max_{q_i}&\  \frac{\xi}{q_i+\hat{q}_{-i}} q_i - \overline{v} q_i - FC-\delta_iq_i\sigma_{v_i}\\
\end{align}
Taking the first order conditions we get:
\begin{align}
0&=\frac{\partial }{\partial q_i}\left(q_i\frac{\xi}{q_i+\hat{q}_{-i}} - q_i\overline{v}-FC-\delta_iq_i\sigma_{v_i}\right)\\
0&= \frac{\xi}{q_i+\hat{q}_{-i}} + \frac{\partial \frac{\xi}{q_i+\hat{q}_{-i}}}{\partial q_i} q_i -(\overline{v}+\delta_i\sigma_{v_i})\\
0&= \frac{\xi}{q_i+\hat{q}_{-i}} - \frac{\xi}{(q_i+\hat{q}_{-i})^2}q_i-(\overline{v}+\delta_i\sigma_{v_i})\\
0&= \xi(q_i + \hat{q}_{-i}) - \xi q_i -(\overline{v}+\delta_i\sigma_{v_i})(q_i+\hat{q}_{-i})^2\\
0&= \frac{\xi \hat{q}_{-i}}{-(\overline{v}+\delta_i\sigma_{v_i})} + q_i^2 + 2\hat{q}_{-i}q_i + \hat{q}_{-i}^2\\
0&= q_i^2 + 2\hat{q}_{-i}q_i + \left( \hat{q}_{-i}^2 - \frac{\xi \hat{q}_{-i}}{(\overline{v}+\delta_i\sigma_{v_i})}\right)
\end{align}

Using the quadratic formula we get:
$$q_i = -\hat{q}_{-i} \pm \frac{1}{2} \sqrt{4\hat{q}_{-i}^2 -4\left( \hat{q}_{-i}^2 - \frac{\xi \hat{q}_{-i}}{(\overline{v}+\delta_i\sigma_{v_i})}\right)}$$
$$q_i = -\hat{q}_{-i} \pm \frac{1}{2} \sqrt{\frac{4\xi \hat{q}_{-i}}{\overline{v}+\delta_i\sigma_{v_i}}}$$
$$q_i = -\hat{q}_{-i} \pm  \sqrt{\frac{\xi \hat{q}_{-i}}{\overline{v}+\delta_i\sigma_{v_i}}}$$
Since each production must be positive and nonzero, it is clear that the only solution is:
$$q_i = -\hat{q}_{-i} +  \sqrt{\frac{\xi \hat{q}_{-i}}{\overline{v}+\delta_i\sigma_{v_i}}}$$
This is the best response equation for each firm assuming the other firm produces at a level of $\hat{q}_{-i}$. Under the Nash equilibrium each firm correctly guessed the other firms production and produces at their best response. Therefore, $\hat{q}_{-i} = q_{-i}^*$ and $\hat{q}_{i} = q_{i}^*$:


\begin{align}
q_i^* &= -\left( -q_{i}^* +  \sqrt{\frac{\xi q_{i}^*}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}}\right) + \sqrt{\frac{\xi \left( -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}}\right)}{\overline{v}+\delta_i\sigma_{v_i}}}\\
q_i^* &= q_{i}^* -  \sqrt{\frac{\xi q_{i}^*}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}} + \sqrt{\frac{\xi \left( -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}}\right)}{\overline{v}+\delta_i\sigma_{v_i}}}\\
\sqrt{\frac{\xi q_{i}^*}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}} &= \sqrt{\frac{\xi \left( -q_{i}^* + \sqrt{\frac{\xi q_{i}^*}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}}\right)}{\overline{v}+\delta_i\sigma_{v_i}}}\\
\frac{1}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}}q_{i}^* &= \frac{-1}{\overline{v}+\delta_{i}\sigma_{v_{i}}}q_{i}^* +\frac{\sqrt{\frac{\xi q_{i}^*}{\overline{v} + \delta_{-i}\sigma_{v_{-i}}}}}{\overline{v}+\delta_{i}\sigma_{v_{i}}}\\
q_{i}*^2\left( \frac{1}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}} + \frac{1}{\overline{v}+\delta_{i}\sigma_{v_{i}}}\right)^2 &= \frac{\xi q_{i}^*}{(\overline{v} + \delta_{-i}\sigma_{v_{-i}})^2(\overline{v} + \delta_{-i}\sigma_{v_{-i}})}\\
0 &= q_{i}*^2\left( \frac{1}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}} + \frac{1}{\overline{v}+\delta_{i}\sigma_{v_{i}}}\right)^2 - \frac{\xi q_{i}^*}{(\overline{v} + \delta_{-i}\sigma_{v_{-i}})^2(\overline{v} + \delta_{-i}\sigma_{v_{-i}})}\\
0 &= q_{i}* \left(\underbrace{\left( \frac{1}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}} + \frac{1}{\overline{v}+\delta_{i}\sigma_{v_{i}}}\right)^2q_{i}* - \frac{\xi}{(\overline{v} + \delta_{-i}\sigma_{v_{-i}})^2(\overline{v} + \delta_{-i}\sigma_{v_{-i}})}}_{=0}\right)\\
q_{i}* &= \frac{\frac{\xi}{(\overline{v} + \delta_{-i}\sigma_{v_{-i}})^2(\overline{v} + \delta_{-i}\sigma_{v_{-i}})}}{\left( \frac{1}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}} + \frac{1}{\overline{v}+\delta_{i}\sigma_{v_{i}}}\right)^2}
\end{align}

As explained in the assumptions, production levels of 0 imply an infinite price and will not be considered. Therefore the solution above shows what each firm produces in the Nash Equilibrium. 
 

## Interpret
We can check this our Nash Equilibrium result by assuming $\delta_i = \delta_{-i}=0$:
\begin{align}
q_i^* &= \frac{\frac{\xi}{\overline{v}^3}}{\left(\frac{1}{\overline{v}}+ \frac{1}{\overline{v}}\right) ^2}\\
&= \frac{\frac{\xi}{\overline{v}^3}}{\left(\frac{2}{\overline{v}}\right)^2}\\
&= \frac{\frac{\xi}{\overline{v}^3}}{\frac{4}{\overline{v}^2}}\\
q_i^* &= \frac{\xi}{4\overline{v}}\\
\end{align}
We get the same result as when firms were risk intolerant.

We will now take partial derivitives to see see how the Nash equilibrium output of our firms changes with changes in their risk tolerance, the other firms risk tolerance, $\xi$, the mean variable cost, the uncertainty in their variable cost, and the uncertianty in the other firms variable cost. 

$$\frac{\partial q_i^*}{\partial \xi} = \frac{\frac{1}{(\overline{v} + \delta_{-i}\sigma_{v_{-i}})^2(\overline{v} + \delta_{-i}\sigma_{v_{-i}})}}{\left( \frac{1}{\overline{v}+\delta_{-i}\sigma_{v_{-i}}} + \frac{1}{\overline{v}+\delta_{i}\sigma_{v_{i}}}\right)^2}$$

It is clear that, since every variable is positive, that partial derivitive is positive. This makes sense becauase a larger $\xi$ means the price is higher for each combination of outputs quantities and the firms output should be higher to capture more revenue by producing more commercial airplanes.



$$\frac{\partial q_i^*}{\partial \delta_i} = \frac{-2\xi\sigma_{v_i}(\overline{v}+\delta_{-i}\sigma_{v_{-i}})}{(\overline{v}+\sigma_{v_i}\delta_i + \delta_{-i}\sigma_{v_{-i}})^3}$$

It is clear that, since every variable is positive, that partial derivitive is negative. This makes sense because as a firm becomes more risk intolerant they will reduce their output to decrease their uncertainty in their profit which is $q_i\sigma_{v_i}$.

Therefore, assuming that the firms have the same uncertainty in their variable cost, which is likely as they are producing essentially the same product, the more risk intolerent a firm is the less they produce and therefore the less profit they capture on average. Being risk tolerant increases a firms expected profit.


$$\frac{\partial q_i^*}{\partial \delta_{-i}} =\frac{-\xi\sigma_{v_{-i}}(-\delta_{i}\sigma_{v_{i}}+\delta_{-i}\sigma_{v_{-i}})}{(\overline{v}+\sigma_{v_i}\delta_i + \delta_{-i}\sigma_{v_{-i}})^3}$$

As $\delta_{i}\sigma_{v_{i}}$ is weighting the uncertianty in the variable cost by firm $i$'s risk intolerance, we will call it the current cost of uncertanty in the variable cost. If the current cost of uncertainty for Firm $i$ is greater than the other firms, then as the other firm becomes more risk intolerant Firm $i$ produces more as the other firm producing less raising the price per product, allowing Firm $i$ to capture more revenue. However, interestingly, if the current cost of uncertainty is less for Firm $i$ than for the other firm, then this term is negative, meaning as the other firm becomes more risk intolerant Firm i produces less. This is because Firm i was already producing less then the other firm and therefore produces even less because the other firm's decrease in production level causes more uncertainty for Firm $i$'s profit than the benefit of the higher price and revenue from the other firms decrease in production. 

$$\frac{\partial q_i^*}{\partial \overline{v}} =\frac{-\xi(2\overline{v}-\delta_{i}\sigma_{v_{i}}+3\delta_{-i}\sigma_{v_{-i}})}{(\overline{v}+\sigma_{v_i}\delta_i + \delta_{-i}\sigma_{v_{-i}})^3}$$

Since as the uncertainty in the variable cost is significantly smaller than the mean variable cost for each firm, this partial derivitive is clearly negative. This means that as the average variable cost increases, the equilibrium output decreases. This makes sense because as mean variable cost increases, the expected profit decreases when you produce the same and the firm could capture more profit by producing less at a higher price. 

$$\frac{\partial q_i^*}{\partial \sigma_{v_i}} =\frac{-2\xi(\overline{v}+\delta_{-i}\sigma_{v_{-i}})}{(\overline{v}+\sigma_{v_i}\delta_i + \delta_{-i}\sigma_{v_{-i}})^3}$$

This partial derivitive is negative. This means that as the uncertainty in a firm's variable cost increases, its equilibrium output decreases. This makes sense as firms wish to lower the uncertainty in their profit and an increase in uncertainty in the variable cost increases the uncertainty in their profit. Therefore, to decrease this uncertainty, $q_i\sigma_{v_i}$, the firm can produce less. 

$$\frac{\partial q_i^*}{\partial \sigma_{v_{-i}}} =\frac{-\xi(-\delta_{i}\sigma_{v_{i}}+\delta_{-i}\sigma_{v_{-i}})}{(\overline{v}+\sigma_{v_i}\delta_i + \delta_{-i}\sigma_{v_{-i}})^3}$$

If the current cost of uncertainty for Firm $i$ is less than the other firms, then as the uncertainty in the variable cost of the other firm increases, Firm $i$ produces more as the other firm producing less, to decrease their profit uncertainty, raising the price per product, allowing Firm $i$ to capture more revenue. However, interestingly, if the current cost of uncertainty is more for Firm $i$ is larger than the other firm's, then this term is negative, and Firm $i$ produces less as the uncertainty in the other firms variable cost increases. This is because Firm $i$ was already producing less then the other firm and therefore produces even less because the other firms decrease in production level causes more uncertainty for Firm $i$'s profit than the benefit of the higher price and revenue from the other firms decrease in production. 



## Conjecture

To make this model better I could have done a number of things. Perhaps I could've assumed that the uncertainty in the variable cost was the same for each firm. This would've simplified outcomes a lot and allowed for clearer intuition. This is also very likely as the firms are producing similar products and face similar production risks. In addition, I could've allowed for uncertianty in the demand curve, through the price elasticity, as consumer choices can easily change, especially for Airbus and Boeing.
